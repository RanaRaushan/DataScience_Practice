# Matplotlib

## Part1

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title("LARGER PLOT")
axes2.set_title("SMALLER PLOT")
```

## -------------------------------Part2-------------------------------

```python
fig, axes = plt.subplots(nrows=1,ncols=2)


# for current_ax in axes:
#     current_ax.plot(x,y)

# or
axes[0].plot(x,y)
axes[0].set_title("First Plot")

axes[1].plot(x,y)
axes[1].set_title("Second Plot")

plt.tight_layout()
```

```python
axes
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig("My_picture.png", dpi =200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label="X Sq")
ax.plot(x,x**3, label="X Cube")

ax.legend()
```

# -------------------------------Part3-------------------------------

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color="purple",linewidth=3, alpha=0.5) # "#FF8C00" RGB Hex Code or provide string color name
ax.plot(x,x**2.2,color="Red",lw=3, linestyle='--')
ax.plot(x,x**2.4,color="Green",lw=3, ls='-.', marker="<", markersize=10)

ax.plot(x,x**2.6,color="blue",lw=3, ls=':', marker="o", ms=7) # linewidth aka lw, linestyle aka ls, markersize aka ms

# markerfacecolor mfc lw, markeredgewidth mew ls, markeredgecolor aka mec
ax.plot(x,x**2.8,color="blue",lw=3, ls=':', marker="o", ms=7, mfc="Yellow", mew=2, mec="red")
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2.2,color="Red",lw=3, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,1])
```
